### simpler-basic

```python
scene_name = "simple"
with open("./log/simple.pkl", "rb") as handle:
    simple_res = pickle.load(handle)
    
scene_name = "basic"
with open("./log/basic.pkl", "rb") as handle:
    basic_res = pickle.load(handle)
    
scene_name = "rocket-basic"
with open("./log/rocket-basic.pkl", "rb") as handle:
    rocket_res = pickle.load(handle)
```

```python
f, axarr = plt.subplots(1,3, figsize=(30,10))
axarr[0].plot(simple_res["loss"])
axarr[0].set_title("simple basic")
axarr[0].set_xlabel("episode")
axarr[0].set_ylabel("training loss")
axarr[1].plot(basic_res["loss"])
axarr[1].set_title("basic")
axarr[1].set_xlabel("episode")
axarr[1].set_ylabel("training loss")
axarr[2].plot(rocket_res["loss"])
axarr[2].set_title("rocket basic")
axarr[2].set_xlabel("episode")
axarr[2].set_ylabel("training loss")
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### simpler-basic

```python
scene_name = "simple"
with open("./log/simple.pkl", "rb") as handle:
    res = pickle.load(handle)
```

```python
plt.plot(res["loss"][:])
plt.title(scene_name)
plt.xlabel("episode")
plt.ylabel("training loss")
plt.show()
```

```python
plt.plot(res["reward"][:])
plt.title(scene_name)
plt.xlabel("episode")
plt.ylabel("reward")
plt.show()
```
